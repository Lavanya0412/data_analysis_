**LINEAR REGRESSION**

```python
#Importing Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()#to save the model
df=pd.read_csv('/content/Salary.csv')
X=df['YearsExperience'].values.reshape(-1,1)
y=df['Salary'].values.reshape(-1,1)
LR.fit(X,y)
```

```python
i = input("Enter ypur experiernce in years: ")
i = float(i)
pred = LR.predict([[i]])#Predicting the values
print("Your predicted Salary is: ",pred[0][0],"INR")
```

```python
import matplotlib.pyplot as plt
yearsexperienced=df['YearsExperience']
salary=df['Salary']
plt.plot(yearsexperienced,salary,color="b")
plt.xlabel('yearsexperienced')
plt.ylabel('salary')
plt.xticks(rotation=45)
plt.show()

```

**LOGISTIC REGRESSION**

```python
#Importing the libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/content/grades(1).csv")
df
```

```python
sem1= df["Curricular units 1st sem (grade)"].values
sem2= df["Curricular units 2nd sem (grade)"].values
avg= df["average score"].values
```

```python
threshold=7
binary_labels=(avg>threshold).astype(int)
```

```python
if avg.ndim == 1:
    avg = np.expand_dims(avg, axis=1)
logr=LogisticRegression()
logr.fit(avg,binary_labels)
```

```python
m=float(input("Enter grade of sem1:"))
n=float(input("Enter grade of sem2:"))
aver=(m+n)/2
predict=logr.predict([[aver]])
print(predict)
if predict==[1]:
  print("Graduated Successfully")
else:
  print("Fail")
```
